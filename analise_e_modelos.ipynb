# Anlise de Dados Categricos

### Primeiramente vamos analizar dentro de cada emprego como  a desigualdade salarial total entre os gneros e se existem reas que so predominantemente ocupadas por homens ou mulheres. 

Sinta-se a vontade para olhar apenas as profisses que lhe sejam interessantes.

```python
Jobs = data['JobTitle'].tolist()
Jobs = list(set(Jobs))
maior_desigualdade=menor_desigualdade=0
for i in Jobs:
    print('')
    print(f'Grfico distribuio de salrio por gnero na profisso {i}:')
    print('')
    print('Mulheres em laranja')
    print('Homens em azul')
    
    data_job=data[data['JobTitle']== i]
    soma_h=data_job[data_job['Gender']=='Male']['BasePay']+data_job[data_job['Gender']=='Male']['Bonus']
    soma_m=data_job[data_job['Gender']=='Female']['BasePay']+data_job[data_job['Gender']=='Female']['Bonus']
    sns.distplot(soma_h, bins=18, hist=True)
    sns.distplot(soma_m, bins=18, hist=True)
    plt.show()
    
    data_job_f=data_job[data_job['Gender']=='Female']
    data_job_m=data_job[data_job['Gender']=='Male']
    print(f'Nmero de mulheres = {len(data_job_f.index)}, nmero de homens = {len(data_job_m.index)}')
    print(f"Mdia salarial mulheres = {int(data_job_f['BasePay'].mean())} Dlares")
    print(f"Mdia salarial homens = {int(data_job_m['BasePay'].mean())} Dlares")
    print(f"Homens ganham {int(data_job_m['BasePay'].mean()-data_job_f['BasePay'].mean())} Dlares a mais do que as mulheres")
    #daqui pra baixo vou separar qual o emprego que tem a maior desigualdade e qual tem a menor para analizarmos depois
    a=data_job_f['BasePay'].mean()
    b=data_job_m['BasePay'].mean()
    c=0
    if a!=0 and b!=0:
        c=b-a
        if c>maior_desigualdade:
            maior_desigualdade=c
            jobinjusto=i
        if c<menor_desigualdade:
            menor_desigualdade=c
            jobjusto=i
    print('')
    print('-------------------------------------------------------------------------------------------------------')
```



Vamos analisar separadamente os trabalhos mais e menos desiguais:

```python
print(f'O trabalho que  mais justo com as mulheres  o de {jobjusto}, com diferena salarial entre homens e mulheres de {int(menor_desigualdade)} Dlares.')
print('')
print(f'Enquanto isso o trabalho menos justo com as mulheres  o de {jobinjusto}, l as mulheres ganham {int(maior_desigualdade)} Dlares a menos que homens.')
data_job_justo=data[data['JobTitle']== jobjusto]
data_job_injusto=data[data['JobTitle']== jobinjusto]
data_job_justo.plot(x='BasePay', y='Gender', alpha=.5, kind='scatter', title= f'Salrios do emprego mais justo, {jobjusto}')
plt.show()
data_job_injusto.plot(x='BasePay', y='Gender', alpha=.5, kind='scatter', title= f'Salrios do emprego mais injusto, {jobinjusto}')
plt.show()
```

 isso mesmo as mulheres ganham mais do que os homens na rea de cincia de dados e elas so muito bem representadas com 54 homens e 53 mulheres! 

Perceba que no apenas as mulheres ganham menos como Engenheiras de Software como tambm so muito raras(8) em proporo aos homens(101)!

### Vamos ver agora se o nvel de escolaridade aumenta ou diminui a desigualdade de renda:

```python
homem_com_HS=dados_homens[dados_homens['Education']== 'High School']['BasePay']
mulher_com_HS= dados_mulheres[dados_mulheres['Education']== 'High School']['BasePay']
sns.distplot(homem_com_HS, bins=18)
sns.distplot(mulher_com_HS, bins=18)
mi_homem_com_HS=homem_com_HS.mean()
mi_mulher_com_HS=mulher_com_HS.mean()

print(f'A diferena salarial entre homens e mulheres com Ensino Mdio ', int(mi_homem_com_HS-mi_mulher_com_HS), 'Dlares.')
print('')
print('Grfico do salrio de homens e mulheres com nvel educacional Ensino Mdio completo:')
print('')
print('Mulheres em laranja')
print('Homens em azul')
```

```python
homem_com_sup= dados_homens[dados_homens['Education']== 'College']['BasePay']
mulher_com_sup= dados_mulheres[dados_mulheres['Education']== 'College']['BasePay']
sns.distplot(homem_com_sup, bins=18)
sns.distplot(mulher_com_sup, bins=18)
mi_homem_com_sup=homem_com_sup.mean()
mi_mulher_com_sup=mulher_com_sup.mean()
print(f'A diferena salarial entre homens e mulheres com graduao ', int(mi_homem_com_sup-mi_mulher_com_sup),'Dlares')
print('')
print('Grfico do salrio de homens e mulheres com nvel educacional Graduao completa:')
print('')
print('Mulheres em laranja')
print('Homens em azul')
```

```python
homem_com_PhD=dados_homens[dados_homens['Education']== 'PhD']['BasePay']
mulher_com_PhD= dados_mulheres[dados_mulheres['Education']== 'PhD']['BasePay']
sns.distplot(homem_com_PhD, bins=18)
sns.distplot(mulher_com_PhD, bins=18)
mi_homem_com_PhD=homem_com_PhD.mean()
mi_mulher_com_PhD=mulher_com_PhD.mean()

print(f'A diferena salarial entre homens e mulheres com Doutorado ', int(mi_homem_com_PhD-mi_mulher_com_PhD),'Dlares')
print('')
print('Grfico do salrio de homens e mulheres com nvel educacional Doutorado:')
print('')
print('Mulheres em laranja')
print('Homens em azul')
```

Como  possvel ver, aps a sada do Ensino mdio a Desigualdade no  muito acentuada, mas a partir da Graduao, o "Gap" aumenta substancialmente e vai diminuindo conforme o nvel acadmico aumenta.

Vejamos a evoluo da Desigualdade Salarial por nveis de senioridade, sendo 1 um inicante e 5 algum no final da carreira.

```python
for i in range(1,6):
    Senioridade_Homem=dados_homens[dados_homens['Seniority']== i]['BasePay']
    Senioridade_Mulher=dados_mulheres[dados_mulheres['Seniority']== i]['BasePay']
    sns.distplot(Senioridade_Homem, bins=18)
    sns.distplot(Senioridade_Mulher, bins=18)
    mi_senioridade_hom=Senioridade_Homem.mean()
    mi_senioridade_mul=Senioridade_Mulher.mean()
    plt.show()
    print(f'A diferena Salarial entre homens e mulheres no nvel {i} de senioridade na carreira  {int(mi_senioridade_hom-mi_senioridade_mul)} Dlares.')
    print('')
    print('')
    print('Mulheres em laranja')
    print('Homens em azul')
```

Como  possvel observar, as mulheres iniciam suas carreiras ganhando menos que os homens, indicando que a desigualdade no necessariamente vem de caractersticas tcnicas, mas sim pelo gnero em si. Isso se evidencia conforme o nvel de senioridade aumenta os salrios se tornam mais meritocrticos e a desigualdade diminui.

```python
import sklearn
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import r2_score 
from sklearn.model_selection import train_test_split
```

```python
dt = pd.DataFrame(data)
gender=pd.get_dummies(dt['Gender'])
dt=pd.concat([dt, gender], axis=1)
Trampo= pd.get_dummies(dt['JobTitle'])
dt=pd.concat([dt, Trampo], axis=1)
X = dt.iloc[:, [2,3,6,9,11,12,13,14,15,16,17,19,20]]#Selecionei as features Age, PerfEval, Seniority, Gender e o Nome do Cargo
y = dt.iloc[:, [7]] #Selecionei o target como BasePay
X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.3, random_state=0) 
lr=LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)

from sklearn.metrics import r2_score 

R2 = r2_score(y_test,y_pred)
print(R2)
```

A acurcia das previses  de 78%!
